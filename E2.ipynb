```python
import pandas as pd
import numpy as np
import math
from collections import Counter
```

```python
df=pd.read_excel('SASTRA University.xlsx',header=None)
```

```python
df.head()
```

```python
def simple_tokenize(text):
    return text.lower().split()
```

```python
bigram_counts=Counter()
unigram_counts=Counter()
```

```python
for sentence in df[0]:
    tokens=simple_tokenize(sentence)
    for token in tokens:
        unigram_counts[token]+=1
    for i in range(len(tokens)-1):
        bigram=(tokens[i],tokens[i+1])
        bigram_counts[bigram]+=1
N=sum(unigram_counts.values())
```

```python
unigram_counts
```

```python
def chi_2_test(w1,w2,c1,c2,c12):
    a=c12
    b=c1-a
    c=c2-a
    r1=a+b
    r2=N-c
    d=r2-c
    e=a+c
    f=b+d
    e_w1=(r1*e)/N
    e_w2=(r1*f)/N
    e_w3=(r2*e)/N
    e_w4=(r2*f)/N
    sum1=(np.square(a-e_w1))/e_w1 if e_w1>0 else 0
    sum2=(np.square(b-e_w2))/e_w2 if e_w2>0 else 0
    sum3=(np.square(c-e_w3))/e_w3 if e_w3>0 else 0
    sum4=(np.square(d-e_w4))/e_w4 if e_w4>0 else 0
    chi_value=sum1+sum2+sum3+sum4
    return {'Chi Value':chi_value,'Word1':w1,'Word2':w2}
```

```python
def t_test(w1,w2,c1,c2,c12):
    p_w1=c1/N
    p_w2=c2/N
    exp_mean=p_w1*p_w2
    ob_mean=c12/N
    s2=ob_mean
    diff=ob_mean-exp_mean
    v=math.sqrt(s2/N) if N>0 else 1
    t=diff/v if v>0 else 0
    t=round(t,2)
    return {'T value':t,'Word1':w1,'Word2':w2}
```

```python
for bigram,bigram_count in bigram_counts.items():
    w1,w2=bigram
    c1=unigram_counts[w1]
    c2=unigram_counts[w2]
    c12=bigram_count
    chi_result=chi_2_test(w1,w2,c1,c2,c12)
    t_result=t_test(w1,w2,c1,c2,c12)
    if chi_result['Chi Value'] > 3.84:  
        print(f"Chi-Square: {chi_result['Chi Value']:.2f} \t Words: {w1} and {w2} are collocated!") 
    else: 
        print(f"Chi-Square: {chi_result['Chi Value']:.2f} \t Words: {w1} and {w2} are not collocated!") 
 
    t_result = t_test(w1, w2, c1, c2, c12) 
    if t_result['T value'] > 2.76: 
        print(f"t-value: {t_result['T value']:.2f} \t Words: {w1} and {w2} are collocated!") 
    else: 
        print(f"t-value: {t_result['T value']:.2f} \t Words: {w1} and {w2} are not collocated!")
```
